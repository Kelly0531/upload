```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import roc_auc_score

import seaborn as sns
import matplotlib.pyplot as plt
import os 

pd.set_option('max_columns',1000)
pd.set_option('max_row',300)
pd.set_option('display.float_format', lambda x: '%.5f' % x)

path = os.getcwd()
print(path)

pd.set_option('max_columns',1000)
pd.set_option('max_row',300)
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
# 按时间顺序将每个股票前80%的数据作为训练集，20%作为测试集
first_flag = 1
for i in range(1,11):
    print(i)
    dfi = pd.read_csv('Stock_'+str(i)+'.csv')
    dfi.rename(columns={'Unnamed: 0':'Time'}, inplace = True)
    dfi['Stock'] = i
    dfi_train = dfi.iloc[0:int(len(dfi)*0.8),:]
    dfi_test = dfi.iloc[int(len(dfi)*0.8):,:]
    print(len(dfi_train) + len(dfi_test))
    if first_flag == 1:
        df_train = dfi_train
        df_test = dfi_test
        first_flag = 0
    else:
        df_train = pd.concat([df_train,dfi_train],axis=0) #按行拼接
        df_test = pd.concat([df_test,dfi_test],axis=0)
print('训练集长度:'+str(len(df_train)))
print('测试集长度:'+str(len(df_test)))
print('总样本长度:'+str(len(df_train)+len(df_test)))
print('训练集正负样本比:'+str(sum(df_train['TargetFeature'])/len(df_train)))
print('测试集正负样本比:'+str(sum(df_test['TargetFeature'])/len(df_test)))
print('总样本正负样本比:'+str((sum(df_train['TargetFeature'])+sum(df_test['TargetFeature']))/(len(df_train)+len(df_test))))
```

```python
# 各特征的缺失值数量
aa = df_train.isnull().sum()
aa[aa>0].sort_values(ascending=False)
```

```python
# 删除Nan和Inf行
df_train = df_train[~df_train.isin([np.nan, np.inf, -np.inf]).any(1)].dropna()
df_test = df_test[~df_test.isin([np.nan, np.inf, -np.inf]).any(1)].dropna()
# 再查看各特征的缺失值数量
aa = df_train.isnull().sum()
aa[aa>0].sort_values(ascending=False)
```

```python
# 离散变量处理
for c in ['Stock']:
    df_train = pd.concat([df_train,pd.get_dummies(df_train[c],prefix = c)],axis = 1)
    df_train = df_train.drop(c,1)
    df_test = pd.concat([df_test,pd.get_dummies(df_test[c],prefix = c)],axis = 1)
    df_test = df_test.drop(c,1)
```

```python
import copy
X_Train = copy.deepcopy(df_train)
del X_Train['Time']
del X_Train['TargetFeature']
Y_Train = df_train['TargetFeature']

X_Test = copy.deepcopy(df_test)
del X_Test['Time']
del X_Test['TargetFeature']
Y_Test = df_test['TargetFeature']
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import roc_auc_score
```

```python
# CV 
def auc_cv(model,X,y):
    auc = cross_val_score(model, X, y, scoring="roc_auc", cv=5)
    return auc
```

```python
# 二分类器对比
names = ["RF", "GBDT","Xgb"]
models = [RandomForestClassifier(),GradientBoostingClassifier(),XGBClassifier()]
for name, model in zip(names, models):
    score = auc_cv(model, X_Train, Y_Train)
    print("{}: {:.6f}, {:.4f}".format(name,score.mean(),score.std()))
```

```python
# 测试集效果
xgb = XGBClassifier()
xgb.fit(X_Train, Y_Train)
y_pred = xgb.predict(X_Test)
y_predprob = xgb.predict_proba(X_Test)[:,1]
print ("Accuracy : %.4g" % metrics.accuracy_score(Y_Test.values, y_pred))
print ("AUC Score (Train): %f" % metrics.roc_auc_score(Y_Test, y_predprob))
```
